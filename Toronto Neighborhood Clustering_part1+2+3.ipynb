### Following code is meant to scrape de wikipedia page for the neighbourhoods data in toronto using beautiful soup

```python
import requests
from bs4 import BeautifulSoup
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(website_url,'lxml') #make soup out of the wiki page
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'}) #take only the table from the soup
My_table
```

```python
links = My_table.findAll('td') #take only what is neaded from the result, between the "td"s
links
```

```python
#make it a string and put it in a list
data = [] 
for x in links:
    data.append(str(x)) 
```

```python
#make new lists for each collumn; parse the old list and fill the new lists as required
pc=[]
b=[]
n=[]
j=1
for i in range(0, len(data)):
    if j==1:
        pc.append(data[i])
        j+=1
    elif j==2:
        b.append(data[i])
        j+=1
    else:
        n.append(data[i])
        j=1
```

```python
#make a dictionary of the lists and the collumns names, and put the dictionary in a panda dataframe
dict = {'Postal code': pc, 'Borough': b, 'Neighborhood': n}  
    
df = pd.DataFrame(dict) 
df.head()
```

```python
#clean the datafrate of unnecesary characters
df['Postal code'] = df['Postal code'].str.replace(r'<td>', '')
df['Postal code'] = df['Postal code'].str.replace(r'\n</td>', '')
df['Borough'] = df['Borough'].str.replace(r'<td>', '')
df['Borough'] = df['Borough'].str.replace(r'\n</td>', '')
df['Neighborhood'] = df['Neighborhood'].str.replace(r'<td>', '')
df['Neighborhood'] = df['Neighborhood'].str.replace(r'\n</td>', '')
df['Neighborhood'] = df['Neighborhood'].str.replace(r'/', ', ')
df.head()
```

```python
#drop any rows that are not usefull
df = df.drop(df[df['Borough'] == 'Not assigned'].index)
df.reset_index(drop=True, inplace=True)
df.head()
```

### After trying multiple methods to find the latitude and longitude of the neighborhoods (using the geocode and the postal codes or the names of the neighborhood(where i got pretty close), or using another json file) i decided to use the csv file provided as follows

```python
df_location=pd.read_csv('http://cocl.us/Geospatial_data') #get the csv file into a dataframe
```

```python
df_location.head()
```

```python
#for each postal code in df find it in the location dataframe and put its latitude and longitude in a new list respectively 
lat=[]
long=[]
for i in range(0, len(df)):
    p=df.loc[i, 'Postal code']
    for j in range (0, len(df_location)):
        l=df_location.loc[j,'Postal Code']
        if p==l:
            lat.append(df_location.loc[j, 'Latitude'])
            long.append(df_location.loc[j, 'Longitude'])
            break
```

```python
#attach the latitude and longitude lists to the original dataframe
df['Latitude']=lat
df['Longitude']=long
df.head()
```

```python
#this makes a function to get all the venues for each neighborhood
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT = 100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        try :
            results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
            venues_list.append([(
                name, 
                lat, 
                lng, 
                v['venue']['name'], 
                v['venue']['location']['lat'], 
                v['venue']['location']['lng'],  
                v['venue']['categories'][0]['name']) for v in results])
        except:
            print(name, 'BROKE')
            pass

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#uses the function and puts it together in a dataframe
Toronto_venues = getNearbyVenues(names=df['Neighborhood'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                                  )
```

```python
# one hot encoding
Toronto_onehot = pd.get_dummies(Toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Toronto_onehot['Neighborhood'] = Toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Toronto_onehot.columns[-1]] + list(Toronto_onehot.columns[:-1])
Toronto_onehot = Toronto_onehot[fixed_columns]

Toronto_onehot.head()
```

```python
#grouping the venues by neighborhood
Toronto_grouped = Toronto_onehot.groupby('Neighborhood').mean().reset_index()
Toronto_grouped
```

```python
#prints the top 5 venues in each neighborhood
num_top_venues = 5

for hood in Toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = Toronto_grouped[Toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#this is a function that sorsts the venues
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = Toronto_grouped['Neighborhood']

for ind in np.arange(Toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
# set number of clusters
kclusters = 5

Toronto_grouped_clustering = Toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

Toronto_merged = df

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
Toronto_merged = Toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

Toronto_merged.head() # check the last columns!
```

```python
#had some trouble with the k-means clustering and it spits out some NaN cluster labels, which need to be removed
Toronto_merged = Toronto_merged.dropna()
```

```python
#also had some trouble with coloring the labels in the folium map. it didnt like the float cluster labels for some reason, so i turned them to int
Toronto_merged['Cluster Labels']=Toronto_merged['Cluster Labels'].astype(int)
Toronto_merged['Cluster Labels']
```

```python
#map all the neighborhoods and color them by the k-means clustering. 
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Toronto_merged['Latitude'], Toronto_merged['Longitude'], Toronto_merged['Neighborhood'], Toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
